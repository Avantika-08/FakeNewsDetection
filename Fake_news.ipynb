```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_merge=pd.concat([data_fake,data_true],axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data=data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
  text=text.lower()
  text=re.sub('\[.*?\]','',text)       #square brackets and their content
  text=re.sub("\\W"," ",text)          #non- word character and spaces
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('<.*?>+','',text)        #html tags
  text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
  text=re.sub('\n','',text)
  text=re.sub('\w*\d\w*','',text)      #alphanumeric words
  return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr= LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"Text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["Text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)

    print("\n\nPrediction: {}".format(output_label(pred_LR[0])))
```

```python
news_input = input("Enter the news text to test: ")
manual_testing(news_input)
```
